```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np 
url = 'https://raw.githubusercontent.com/yveness7/Marine_Mammals_vs_Climate_Change_mid_term_project/main/mammals%20data/arctic_ice_extent.csv'

ice = pd.read_csv(url)
ice
```

## Calculating rate of change

```python
rate_of_change = lambda x1, y1, x2, y2: (y1 - y2) / (x1 - x2) if x1 != x2 else float('inf')


ice['rate_of_change'] = ice.apply(lambda row: rate_of_change(row['year'], row['extent'], 
                                ice.at[row.name - 1, 'year'] if row.name > 0 else row['year'], 
                                ice.at[row.name - 1, 'extent'] if row.name > 0 else row['extent']), axis=1)

ice
```

## Calculate average rate of change for sea ice per decade

```python
ice['decade'] = (ice['year'] // 10) * 10
#ice['decade']
# Calculating average rate per decade
avg_rate_per_decade = ice.groupby('decade')['rate_of_change'].mean()
avg_rate_per_decade
```
